```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ankit\Desktop\Google DA Certificate\Bellabeat Case Study\Bellabeat Dataset\Fitabase Data 4.12.16-5.12.16\sleepDay_merged.csv")
```

```python
df.head
```

```python
sleepDaydf=df
```

```python
sleepDaydf['TotalMinutesAsleep'] = pd.to_numeric(sleepDaydf['TotalMinutesAsleep'], errors='coerce')
```

```python
pivot_table = pd.pivot_table(
    df, 
    values=['TotalMinutesAsleep', 'TotalTimeInBed'], 
    index='Id', 
    aggfunc='mean', 
    fill_value=0,
    margins=True
)
print(pivot_table)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7, 4))
plt.scatter(df['TotalMinutesAsleep'], df['TotalTimeInBed'])
plt.xlabel('Total Minutes Asleep')
plt.ylabel('Total Time In Bed')
plt.title('Scatter Plot of Total Minutes Asleep vs Total Time In Bed')
plt.grid(True)
plt.show()
```
